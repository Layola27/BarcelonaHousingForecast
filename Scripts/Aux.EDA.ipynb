```python
if not df.empty:
    # Calcula y muestra el porcentaje de valores ausentes por columna
    missing_percentage = (df.isnull().sum() / len(df)) * 100
    missing_df = missing_percentage[missing_percentage > 0].sort_values(ascending=False)

    if not missing_df.empty:
        print("Porcentaje de Valores Ausentes por Columna (solo con ausentes):")
        display(missing_df.to_frame(name='Porcentaje Ausente (%)'))

        # Visualización de valores ausentes
        plt.figure(figsize=(10, len(missing_df) * 0.3 if len(missing_df) > 0 else 3)) # Ajusta altura dinámicamente
        sns.barplot(x=missing_df.values, y=missing_df.index, palette='viridis')
        plt.title('Porcentaje de Valores Ausentes')
        plt.xlabel('Porcentaje (%)')
        plt.show()
    else:
        print("No hay valores ausentes en el dataset.")
else:
    print("DataFrame vacío. No se puede analizar valores ausentes.")
```

```python
if not df.empty:
    # Lista de columnas numéricas relevantes para tu dataset
    numerical_cols = ['price', 'size', 'rooms', 'bathrooms', 'priceByArea', 'antiguedad', 'period_in_market_days']
    # Filtra solo las columnas que existen y son numéricas
    numerical_cols = [col for col in numerical_cols if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]

    print("--- Análisis Univariado: Variables Numéricas ---")
    for col in numerical_cols:
        plt.figure(figsize=(12, 4))
        # Histograma para ver la distribución
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f'Histograma de {col}')

        # Boxplot para identificar outliers y dispersión
        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot de {col}')
        plt.tight_layout()
        plt.show()
        print(f"Estadísticas básicas para '{col}': Media={df[col].mean():.2f}, Mediana={df[col].median():.2f}, Desv.Std={df[col].std():.2f}\n")
else:
    print("DataFrame vacío. No se puede realizar análisis univariado numérico.")
```

```python
if not df.empty:
    # Lista de columnas categóricas relevantes
    categorical_cols = ['propertyType', 'operation', 'distrito', 'barrio', 'hasLift', 'exterior', 'floor', 'nuevo', 'buen_estado', 'a_reformar', 'eficiencia_energetica_tipo']
    # Filtra solo las columnas que existen
    categorical_cols = [col for col in categorical_cols if col in df.columns]

    print("--- Análisis Univariado: Variables Categóricas ---")
    for col in categorical_cols:
        plt.figure(figsize=(10, 5))
        # Para alta cardinalidad, mostrar solo las N más frecuentes
        top_n = 15
        if df[col].nunique() > top_n and col in ['barrio', 'floor', 'agency', 'address']:
            counts = df[col].value_counts().nlargest(top_n)
            sns.barplot(x=counts.index, y=counts.values, palette='viridis', order=counts.index)
            plt.title(f'Top {top_n} Frecuencias en {col}')
        else:
            counts = df[col].value_counts()
            sns.barplot(x=counts.index, y=counts.values, palette='viridis', order=counts.index)
            plt.title(f'Frecuencias en {col}')

        plt.xlabel(col)
        plt.ylabel('Frecuencia')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
        print(f"'{col}' tiene {df[col].nunique()} categorías únicas. Mostrando frecuencias (top 5):")
        display(df[col].value_counts(normalize=True).head().apply(lambda x: f"{x*100:.2f}%"))
        print("-" * 50)
else:
    print("DataFrame vacío. No se puede realizar análisis univariado categórico.")
```

```python
if not df.empty:
    print("--- Análisis Bivariado: Numérico vs. Numérico ---")

    # Scatter plot: Tamaño vs Precio
    if 'size' in df.columns and 'price' in df.columns:
        plt.figure(figsize=(8, 5))
        sns.scatterplot(x='size', y='price', data=df, alpha=0.5)
        plt.title('Tamaño vs. Precio')
        plt.xlabel('Tamaño (m²)')
        plt.ylabel('Precio (€)')
        plt.show()

    # Matriz de Correlación (para un subconjunto de variables numéricas relevantes)
    corr_cols = ['price', 'size', 'rooms', 'bathrooms', 'priceByArea', 'antiguedad', 'latitude', 'longitude']
    corr_cols = [col for col in corr_cols if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]
    if len(corr_cols) > 1:
        correlation_matrix = df[corr_cols].corr()
        plt.figure(figsize=(10, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
        plt.title('Matriz de Correlación (Variables Seleccionadas)')
        plt.show()
else:
    print("DataFrame vacío. No se puede realizar análisis bivariado numérico-numérico.")
```

```python
if not df.empty:
    print("--- Análisis Bivariado: Numérico vs. Categórico ---")

    # Precio vs Distrito (Boxplot)
    if 'price' in df.columns and 'distrito' in df.columns:
        plt.figure(figsize=(12, 6))
        # Ordenar distritos por precio mediano para mejor visualización
        order = df.groupby('distrito')['price'].median().sort_values(ascending=False).index
        sns.boxplot(x='distrito', y='price', data=df, palette='Set2', order=order, showfliers=False) # showfliers=False para ocultar outliers y ver mejor las cajas
        plt.title('Precio por Distrito')
        plt.xticks(rotation=45, ha='right')
        # plt.ylim(0, df['price'].quantile(0.99)) # Descomentar para limitar eje Y si hay outliers extremos
        plt.tight_layout()
        plt.show()

    # Precio vs Tipo de Propiedad (Violinplot)
    if 'price' in df.columns and 'propertyType' in df.columns:
        plt.figure(figsize=(10, 5))
        order = df.groupby('propertyType')['price'].median().sort_values(ascending=False).index
        sns.violinplot(x='propertyType', y='price', data=df, palette='husl', order=order, cut=0) # cut=0 para limitar el violin al rango de datos
        plt.title('Precio por Tipo de Propiedad')
        plt.xticks(rotation=45, ha='right')
        # plt.ylim(0, df['price'].quantile(0.99)) # Descomentar para limitar eje Y
        plt.tight_layout()
        plt.show()
else:
    print("DataFrame vacío. No se puede realizar análisis bivariado numérico-categórico.")
```

```python
if not df.empty:
    print("--- Análisis Bivariado: Categórico vs. Categórico ---")

    # Tipo de Propiedad vs Tiene Ascensor (Countplot)
    if 'propertyType' in df.columns and 'hasLift' in df.columns:
        plt.figure(figsize=(10, 6))
        sns.countplot(x='propertyType', hue='hasLift', data=df, palette='mako')
        plt.title('Tipo de Propiedad vs. Disponibilidad de Ascensor')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
        # Tabla de contingencia (porcentajes)
        display(pd.crosstab(df['propertyType'], df['hasLift'], normalize='index').applymap(lambda x: f"{x*100:.1f}%"))

    # Distrito vs Estado de Conservación (ej. 'buen_estado')
    if 'distrito' in df.columns and 'buen_estado' in df.columns:
        top_distritos = df['distrito'].value_counts().nlargest(7).index # Tomar solo los N distritos más comunes
        df_top_distritos = df[df['distrito'].isin(top_distritos)]
        plt.figure(figsize=(12, 6))
        sns.countplot(x='distrito', hue='buen_estado', data=df_top_distritos, palette='OrRd', order=top_distritos)
        plt.title('Estado de Conservación ("buen_estado") por Distrito (Top 7)')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
else:
    print("DataFrame vacío. No se puede realizar análisis bivariado categórico-categórico.")
```
